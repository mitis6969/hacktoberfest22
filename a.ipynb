```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,1,0.02)
```

```python
y_true=np.sin(1+x**2)
y_noise=y_true+np.random.normal(0,0.032,(50,))
```

```python
plt.scatter(x,y_noise)
plt.scatter(x,y_true)
plt.show()
```

```python
x_train,x_test=x[:40].reshape(40,-1),x[40:].reshape(10,-1)
y_true_train,y_true_test=y_true[:40].reshape(40,-1),y_true[40:].reshape(10,-1)
y_train,y_test=y_noise[:40].reshape(40,-1),y_noise[40:].reshape(10,-1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
mean_squared_error(y_test,pred)
```

```python
plt.plot(x_test,pred)
plt.plot(x_test,y_test)
plt.plot(x_test,y_true_test)
```

```python
from sklearn.pipeline import make_pipeline

poly=make_pipeline(PolynomialFeatures(3),LinearRegression())
poly.fit(x_train,y_train)

pred=poly.predict(x_test)

print(mean_squared_error(y_test,pred))

plt.figure(figsize=(7,7))
plt.plot(x_test,pred)
plt.plot(x_test,y_test)
plt.plot(x_test,y_true_test)
plt.legend(["pred","noisy","true"],loc="upper left")
plt.show()
```
